```python
#Dependencies
import pandas as pd
from sqlalchemy import create_engine
import psycopg2
```

```python
#Read CSV
data_path="data.csv"
data=pd.read_csv(data_path)
data.drop(['type', 'species', 'coat', 'published_at', 'distance', 'search_address', 'search_city', 'search_state'], axis=1, inplace =True)
data.rename(columns={'Unnamed: 0':'dog_id'},inplace=True)
data.head()
```

```python
#Create Database Connection
from config import password
engine = create_engine(f'postgresql+psycopg2://postgres:{password}@localhost/Doghouse')
con = engine.connect()
```

```python
#Load Data Frames into Database
data.to_sql(name='data', con=engine, if_exists='append', index=False)
```

```python
#Confirm Data Frames are in Database
engine.table_names()
```
